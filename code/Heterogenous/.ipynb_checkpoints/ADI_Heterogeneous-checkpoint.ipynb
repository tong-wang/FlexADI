```python
import numpy as np
import pandas as pd
```

## Load data from file

```python
## read the output files
ADI_I = pd.read_csv("Result_ADI_Hetero_I.txt", sep=r"\t")
ADI_II = pd.read_csv("Result_ADI_Hetero_II.txt", sep=r"\t")
ADIF_I = pd.read_csv("Result_ADIF_Simulation_I.txt", sep=r"\t")
ADIF_II = pd.read_csv("Result_ADIF_Simulation_II.txt", sep=r"\t")
## cost in the tradition model without ADI as a benchmark
cost_Traditional = 420.1955945  # taken from Result_ADI_Homo_I_N12.txt
```

## Plots

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

### Figure 7. Costs of the Six Models Under a Typical Setting

```python
typical_index_I = (ADI_I['K'] == 100) & (ADI_I['h'] == 1) & (ADI_I['b'] == 9)
typical_index_II = (ADI_II['K'] == 100) & (ADI_II['h'] == 1) & (ADI_II['b'] == 9)
```

```python
ExpNo_typical_II = 6 - ADI_II.loc[typical_index_II, "lambda0"]
ExpNo_typical_II.name = "ExpNo"
```

```python
cost_ADI_typical_II = ADI_II.loc[typical_index_II, "Cost"]
cost_ADI_typical_II.name = "Cost_ADI"
cost_AP_typical_II = ADIF_II.loc[typical_index_II, "Simu_AP"]
cost_PL_0_typical_II = ADIF_II.loc[typical_index_II, "Simu_PL_0"]
cost_PL_s_typical_II = ADIF_II.loc[typical_index_II, "Simu_PL_s"]
cost_PL_S_typical_II = ADIF_II.loc[typical_index_II, "Simu_PL_S"]
```

```python
# plot Figure 7
fig7 = plt.figure(figsize=(10, 8))
ax = plt.subplot(111)
ax.set_xlim(0.9, 6.1)
ax.set_ylim(260, 430)
ax.set_yticks([260, 300, 340, 380, 420])
ax.set_xlabel("Experiment number", fontsize=14)
ax.set_ylabel("Cost", fontsize=14)
ax.plot((1,6), (cost_Traditional, cost_Traditional), color="black", linestyle='dotted', label="(a) Traditional")
ax.plot(ExpNo_typical_II, cost_ADI_typical_II, color="0.3", linestyle="dashed", marker="x", label="(b) ADI")
ax.plot(ExpNo_typical_II, cost_AP_typical_II, color="0.3", linestyle="dashed", marker="o", label="(c) ADIF AP")
ax.plot(ExpNo_typical_II, cost_PL_0_typical_II, color="0.5", marker="*", label="(d) ADIF PL(0)")
ax.plot(ExpNo_typical_II, cost_PL_s_typical_II, color="0", marker="s", label="(e) ADIF PL($\sigma$)")
ax.plot(ExpNo_typical_II, cost_PL_S_typical_II, color="0.7", marker="D", label="(f) ADIF PL($\Sigma$)")
ax.legend(loc=3, fontsize=14)
plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure7.pdf')
pp.savefig(fig7)
pp.close()
```

## Figure 9. Optimal Costs With and Without Delivery flexibility as Functions of L and ADI Availability

```python
# construct L, ExpNo arrays for 3D plot
Ls = xrange(0, 5)
ExpNos = xrange(1, 7)
L, ExpNo = np.meshgrid(Ls, ExpNos)
```

```python
# merge Case I and Case II into one dataframe
ADI_typical_I = ADI_I.ix[typical_index_I, ["L", "lambda0", "Cost"]].copy().append(ADI_II.ix[typical_index_II, ["L", "lambda0", "Cost"]], ignore_index=True)
ADIF_typical_I = ADIF_I.ix[typical_index_I, ["L", "lambda0", "Simu_PL_s"]].copy().append(ADIF_II.ix[typical_index_I, ["L", "lambda0", "Simu_PL_s"]], ignore_index=True)
```

```python
# add column ExpNo
ADI_typical_I["ExpNo"] = 6 - ADI_typical_I["lambda0"]
ADIF_typical_I["ExpNo"] = 6 - ADIF_typical_I["lambda0"]
```

```python
# extract Cost_ADI and Cost_ADIF (that correspond to the right L and T) from the merged dataframe
fig9_ADI = [[ADI_typical_I.loc[(ADI_typical_I['L'] == l) & (ADI_typical_I['ExpNo'] == exp), "Cost"].iloc[0]  for l in Ls] for exp in ExpNos]
fig9_ADIF = [[ADIF_typical_I.loc[(ADIF_typical_I['L'] == l) & (ADIF_typical_I['ExpNo'] == exp), "Simu_PL_s"].iloc[0]  for l in Ls] for exp in ExpNos]
```

```python
# plot Figure 9
fig9 = plt.figure(figsize=(11, 8))
ax = fig9.gca(projection='3d')
ax.w_xaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_yaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_zaxis.set_pane_color((1, 1, 1, 1.0))
ax.plot_wireframe(L, ExpNo, fig9_ADIF, rstride=1, cstride=1, colors="black")
ax.plot_wireframe(L, ExpNo, fig9_ADI, rstride=1, cstride=1, colors="DimGray", linestyles="--")
ax.scatter(L, T, ADI, s=40, color="gray")
ax.scatter(L, T, ADIF, s=40, color="black")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)
ax.set_zlim(600, 1500)
ax.xaxis.set_major_locator(LinearLocator(5))
ax.yaxis.set_major_locator(LinearLocator(3))
ax.zaxis.set_major_locator(LinearLocator(4))
ax.set_xlabel("L", fontsize=14)
ax.set_ylabel("T", fontsize=14)
ax.set_zlabel("Cost", fontsize=14)
plt.gca().invert_xaxis()
ax.view_init(30, -40)
plt.show()
```
