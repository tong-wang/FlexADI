```python
import numpy as np
import pandas as pd
```

## Load data from file

```python
## read the output files
ADI_I = pd.read_csv("Result_ADI_Homo_I.txt", sep=r"\t")
ADI_II = pd.read_csv("Result_ADI_Homo_II.txt", sep=r"\t")
ADIF_I = pd.read_csv("Result_ADIF_Homo_I.txt", sep=r"\t")
ADIF_II = pd.read_csv("Result_ADIF_Homo_II.txt", sep=r"\t")
```

## Plots

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

### Figure 3. Plot the optimal state-dependent (s(V), S(V)) policy

```python
# construct (U, V, Y) arrays
Vs = xrange(0, 31)
Us = xrange(-100, 51)
U, V = np.meshgrid(Us, Vs)
Y = [[ADIF_II["S" + str(v)][0] if u <= ADIF_II["s" + str(v)][0] else u for u in Us] for v in Vs]
```

```python
# plot Figure 3
fig3 = plt.figure(figsize=(11, 8))
ax = fig3.gca(projection='3d')
ax.w_xaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_yaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_zaxis.set_pane_color((1, 1, 1, 1.0))
ax.plot_wireframe(U, V, Y, rstride=1, cstride=3, colors="DimGray")
ax.set_xlim(-100, 50)
ax.set_ylim(0, 30)
ax.set_zlim(-40, 60)
ax.xaxis.set_major_locator(LinearLocator(4))
ax.yaxis.set_major_locator(LinearLocator(7))
ax.zaxis.set_major_locator(LinearLocator(6))
ax.set_xlabel("$u$", fontsize=18)
ax.set_ylabel("$\hat{v}$", fontsize=18)
ax.set_zlabel("$y$", fontsize=18)
plt.gca().invert_xaxis()
ax.view_init(30, -40)
plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure3.pdf')
pp.savefig(fig3)
pp.close()
```

### Figure 4. Cost comparison between ADI and ADIF

```python
# construct L, T arrays for 3D plot
Ls = xrange(0, 5)
Ts = xrange(0, 3)
L, T = np.meshgrid(Ls, Ts)
```

```python
# merge Case I and Case II into one dataframe
fig4_ADI = ADI_I.ix[:,0:8].copy().append(ADI_II.ix[:, 0:8], ignore_index=True)
fig4_ADIF = ADIF_I.ix[:,0:8].copy().append(ADIF_II.ix[:, 0:8], ignore_index=True)
```

```python
# extract Cost_ADI and Cost_ADIF (that correspond to the right L and T) from the merged dataframe
ADI = [[fig4_ADI.loc[(fig4_ADI['L'] == l) & (fig4_ADI['T'] == t), "Cost"].iloc[0]  for l in Ls] for t in Ts]
ADIF = [[fig4_ADIF.loc[(fig4_ADIF['L'] == l) & (fig4_ADIF['T'] == t), "Cost"].iloc[0]  for l in Ls] for t in Ts]
```

```python
# plot Figure 4
fig4 = plt.figure(figsize=(11, 8))
ax = fig4.gca(projection='3d')
ax.w_xaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_yaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_zaxis.set_pane_color((1, 1, 1, 1.0))
ax.plot_wireframe(L, T, ADIF, rstride=1, cstride=1, colors="black")
ax.plot_wireframe(L, T, ADI, rstride=1, cstride=1, colors="DimGray", linestyles="--")
ax.scatter(L, T, ADI, s=40, color="gray")
ax.scatter(L, T, ADIF, s=40, color="black")
ax.set_xlim(0, 4)
ax.set_ylim(0, 2)
ax.set_zlim(600, 1500)
ax.xaxis.set_major_locator(LinearLocator(5))
ax.yaxis.set_major_locator(LinearLocator(3))
ax.zaxis.set_major_locator(LinearLocator(4))
ax.set_xlabel("L", fontsize=14)
ax.set_ylabel("T", fontsize=14)
ax.set_zlabel("Cost", fontsize=14)
plt.gca().invert_xaxis()
ax.view_init(30, -40)
plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure4.pdf')
pp.savefig(fig4)
pp.close()
```
