```python
import numpy as np
import pandas as pd
```

## Load data from file

```python
## read the output files
ADI_Hetero_I = pd.read_csv("Result_ADI_Hetero_I.txt", sep=r"\t")
ADI_Hetero_II = pd.read_csv("Result_ADI_Hetero_II.txt", sep=r"\t")
ADIF_Hetero_I = pd.read_csv("Result_ADIF_Simulation_I.txt", sep=r"\t")
ADIF_Hetero_II = pd.read_csv("Result_ADIF_Simulation_II.txt", sep=r"\t")
```

```python
# add ExpNo to dataframes
ADI_Hetero_I["ExpNo"] = 6 - ADI_Hetero_I["lambda0"]
ADI_Hetero_II["ExpNo"] = 6 - ADI_Hetero_II["lambda0"]
ADIF_Hetero_I["ExpNo"] = 6 - ADIF_Hetero_I["lambda0"]
ADIF_Hetero_II["ExpNo"] = 6 - ADIF_Hetero_II["lambda0"]
```

```python
# merge all data from Case I and II
ADI_Hetero = ADI_Hetero_I.ix[:, ["L", "ExpNo", "K", "h", "b", "Cost"]].copy().append(ADI_Hetero_II.ix[:, ["L", "ExpNo", "K", "h", "b", "Cost"]], ignore_index=True)
ADIF_Hetero = ADIF_Hetero_I.ix[:, ["L", "ExpNo", "K", "h", "b", "Simu_AP", "Simu_PL_0", "Simu_PL_s", "Simu_PL_S"]].copy().append(ADIF_Hetero_II.ix[:, ["L", "ExpNo", "K", "h", "b", "Simu_AP", "Simu_PL_0", "Simu_PL_s", "Simu_PL_S"]], ignore_index=True)
```

## Plots

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib import gridspec
```

### Figure 7. Costs of the Six Models Under a Typical Setting

The typical setting: K=100, h=1, b=9. Case II: T=2, L=0.

```python
## cost in the tradition model without ADI as a benchmark
cost_Traditional_typical = 420.1955945  # taken from Result_ADI_Homo_I_N12.txt
```

```python
# index for subsetting the typical case
typical_index = (ADI_Hetero['L'] == 0) & (ADI_Hetero['K'] == 100) & (ADI_Hetero['h'] == 1) & (ADI_Hetero['b'] == 9)
```

```python
# extract costs of the typical case
cost_ADI_typical = ADI_Hetero.loc[typical_index, "Cost"]
cost_AP_typical = ADIF_Hetero.loc[typical_index, "Simu_AP"]
cost_PL_0_typical = ADIF_Hetero.loc[typical_index, "Simu_PL_0"]
cost_PL_s_typical = ADIF_Hetero.loc[typical_index, "Simu_PL_s"]
cost_PL_S_typical = ADIF_Hetero.loc[typical_index, "Simu_PL_S"]
```

```python
# extract x-axis of the typical case
ExpNos = ADIF_Hetero.loc[typical_index, "ExpNo"]
```

```python
# plot Figure 7
fig7 = plt.figure(figsize=(10, 8))
ax = plt.subplot(111)
ax.set_xlim(0.9, 6.1)
ax.set_ylim(260, 430)
ax.set_yticks([260, 300, 340, 380, 420])
ax.set_xlabel("Experiment number", fontsize=14)
ax.set_ylabel("Cost", fontsize=14)
ax.plot((1,6), (cost_Traditional_typical, cost_Traditional_typical), color="black", linestyle='dotted', label="(a) Traditional")
ax.plot(ExpNos, cost_ADI_typical, color="0.3", linestyle="dashed", marker="x", label="(b) ADI")
ax.plot(ExpNos, cost_AP_typical, color="0.3", linestyle="dashed", marker="o", label="(c) ADIF AP")
ax.plot(ExpNos, cost_PL_0_typical, color="0.5", marker="*", label="(d) ADIF PL(0)")
ax.plot(ExpNos, cost_PL_s_typical, color="0", marker="s", label="(e) ADIF PL($\sigma$)")
ax.plot(ExpNos, cost_PL_S_typical, color="0.7", marker="D", label="(f) ADIF PL($\Sigma$)")
ax.legend(loc=3, fontsize=14)
plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure7.pdf')
pp.savefig(fig7)
pp.close()
```

### Figure 8. Optimality Gaps of the Heuristics (Subtotaled Average)

```python
fig8_data = ADIF_Hetero.ix[(ADIF_Hetero['ExpNo'] >= 2) & (ADIF_Hetero['ExpNo'] <= 5)]
```

```python
# calculate delta's
fig8_data["delta_0"] = 100 * (fig8_data["Simu_PL_0"] - fig8_data["Simu_AP"]) / fig8_data["Simu_AP"]
fig8_data["delta_s"] = 100 * (fig8_data["Simu_PL_s"] - fig8_data["Simu_AP"]) / fig8_data["Simu_AP"]
fig8_data["delta_S"] = 100 * (fig8_data["Simu_PL_S"] - fig8_data["Simu_AP"]) / fig8_data["Simu_AP"]
```

```python
fig8_data.head()
```

```python
# aggregation by ExpNo, h, and b
fig8_ExpNo = fig8_data.groupby('ExpNo').aggregate(np.mean).loc[:, ["delta_0", "delta_s", "delta_S"]]
fig8_h = fig8_data.groupby('h').aggregate(np.mean).loc[:, ["delta_0", "delta_s", "delta_S"]]
fig8_b = fig8_data.groupby('b').aggregate(np.mean).loc[:, ["delta_0", "delta_s", "delta_S"]]
```

```python
# plot Figure 8
fig8 = plt.figure(figsize=(16, 4))
gs = gridspec.GridSpec(1, 3, width_ratios=[5, 3, 3]) 

bar_width1 = 0.25
ax1 = plt.subplot(gs[0])
ax1.bar(fig8_ExpNo.index.values - bar_width1, fig8_ExpNo['delta_s'], bar_width1, color="0.2", label="$\delta_\sigma$")
ax1.bar(fig8_ExpNo.index.values, fig8_ExpNo['delta_S'], bar_width1, color="0.5", label="$\delta_\Sigma$")
ax1.bar(fig8_ExpNo.index.values + bar_width1, fig8_ExpNo['delta_0'], bar_width1, color="0.8", label="$\delta_0$")
ax1.set_xlim(0.5, 6.5)
ax1.set_ylim(0, 8)
ax1.set_xticks([1, 2, 3, 4, 5, 6])
ax1.set_yticks([0, 2, 4, 6, 8])
ax1.set_xlabel("Experiment number")
ax1.set_ylabel("Optimality gap (%)")
ax1.legend(loc=2, fontsize=14)

bar_width2 = 0.4
ax2 = plt.subplot(gs[1])
ax2.bar(fig8_h.index.values - 1.5*bar_width2, fig8_h['delta_s'], bar_width2, color="0.2", label="$\delta_\sigma$")
ax2.bar(fig8_h.index.values - 0.5*bar_width2, fig8_h['delta_S'], bar_width2, color="0.5", label="$\delta_\Sigma$")
ax2.bar(fig8_h.index.values + 0.5*bar_width2, fig8_h['delta_0'], bar_width2, color="0.8", label="$\delta_0$")
ax2.set_xlim(0, 6)
ax2.set_ylim(0, 8)
ax2.set_xticks([1, 3, 5])
ax2.set_yticks([0, 2, 4, 6, 8])
ax2.set_xlabel("h")

bar_width3 = 2
ax3 = plt.subplot(gs[2])
ax3.bar(fig8_b.index.values - 1.5*bar_width3, fig8_b['delta_s'], bar_width3, color="0.2", label="$\delta_\sigma$")
ax3.bar(fig8_b.index.values - 0.5*bar_width3, fig8_b['delta_S'], bar_width3, color="0.5", label="$\delta_\Sigma$")
ax3.bar(fig8_b.index.values + 0.5*bar_width3, fig8_b['delta_0'], bar_width3, color="0.8", label="$\delta_0$")
ax3.set_xlim(4, 34)
ax3.set_ylim(0, 8)
ax3.set_xticks([9, 19, 29])
ax3.set_yticks([0, 2, 4, 6, 8])
ax3.set_xlabel("p")

plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure8.pdf')
pp.savefig(fig8)
pp.close()
```

### Figure 9. Optimal Costs With and Without Delivery flexibility as Functions of L and ADI Availability

```python
# construct L, ExpNo arrays for 3D plot
Ls = xrange(0, 5)
ExpNos = xrange(1, 7)
L, ExpNo = np.meshgrid(Ls, ExpNos)
```

```python
# index for subsetting
typical_index_fig9 = (ADI_Hetero['K'] == 100) & (ADI_Hetero['h'] == 1) & (ADI_Hetero['b'] == 9)
```

```python
ADI_typical = ADI_Hetero.loc[typical_index_fig9]
ADIF_typical = ADIF_Hetero.loc[typical_index_fig9]
```

```python
# extract Cost_ADI and Cost_ADIF (that correspond to the right L and T)
fig9_ADI = [[ADI_typical.loc[(ADI_typical['L'] == l) & (ADI_typical['ExpNo'] == exp), "Cost"].iloc[0]  for l in Ls] for exp in ExpNos]
fig9_ADIF = [[ADIF_typical.loc[(ADIF_typical['L'] == l) & (ADIF_typical['ExpNo'] == exp), "Simu_PL_s"].iloc[0]  for l in Ls] for exp in ExpNos]
```

```python
# plot Figure 9
fig9 = plt.figure(figsize=(11, 8))
ax = fig9.gca(projection='3d')
ax.w_xaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_yaxis.set_pane_color((1, 1, 1, 1.0))
ax.w_zaxis.set_pane_color((1, 1, 1, 1.0))
ax.plot_wireframe(L, ExpNo, fig9_ADIF, rstride=1, cstride=1, colors="0.1")
ax.plot_wireframe(L, ExpNo, fig9_ADI, rstride=1, cstride=1, colors="0.4", linestyle="dashed")
ax.scatter(L, ExpNo, fig9_ADIF, s=40, color="0.1", marker="*")
ax.scatter(L, ExpNo, fig9_ADI, s=40, color="0.4")
ax.set_xlim(0, 4)
ax.set_ylim(1, 6)
ax.set_zlim(250, 800)
ax.xaxis.set_major_locator(LinearLocator(5))
ax.yaxis.set_major_locator(LinearLocator(6))
ax.set_zticks([300, 400, 500, 600, 700, 800])
ax.set_xlabel("L", fontsize=14)
ax.set_ylabel("Experiment number", fontsize=14)
ax.set_zlabel("Cost", fontsize=14)
plt.gca().invert_xaxis()
ax.view_init(30, -40)
plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure9.pdf')
pp.savefig(fig9)
pp.close()
```

### Figure 10. Cost Savings of Delivery Flexibility (Subtotaled Average)

```python
fig10_data = ADI_Hetero.ix[:, 0:5]
fig10_data["Delta"] = 100 * (ADI_Hetero["Cost"] - ADIF_Hetero["Simu_PL_s"]) / ADI_Hetero["Cost"]
fig10_data["Delta_bar"] = 100 * (ADI_Hetero["Cost"] - ADIF_Hetero["Simu_AP"]) / ADI_Hetero["Cost"]
```

```python
fig10_data.head()
```

```python
# aggregation by ExpNo, h, and b
fig10_ExpNo = fig10_data.groupby('ExpNo').aggregate(np.mean).loc[:, ["Delta", "Delta_bar"]]
fig10_h = fig10_data.groupby('h').aggregate(np.mean).loc[:, ["Delta", "Delta_bar"]]
fig10_b = fig10_data.groupby('b').aggregate(np.mean).loc[:, ["Delta", "Delta_bar"]]
```

```python
# plot Figure 10
fig10 = plt.figure(figsize=(16, 4))
gs = gridspec.GridSpec(1, 3, width_ratios=[6, 3, 3]) 

bar_width1 = 0.6
ax1 = plt.subplot(gs[0])
ax1.bar(fig10_ExpNo.index.values - 0.5*bar_width1, fig10_ExpNo['Delta_bar'], bar_width1, color="0.8", label="$\Delta$")
ax1.bar(fig10_ExpNo.index.values - 0.5*bar_width1, fig10_ExpNo['Delta'], bar_width1, color="0.4", label="$\Delta$")
ax1.set_xlim(0.5, 6.5)
ax1.set_ylim(0, 30)
ax1.set_xticks([1, 2, 3, 4, 5, 6])
ax1.set_xlabel("Experiment number")
ax1.set_ylabel("Cost savings (%)")
#ax1.legend(loc=2, fontsize=14)

bar_width2 = 1.2
ax2 = plt.subplot(gs[1])
ax2.bar(fig10_h.index.values - 0.5*bar_width2, fig10_h['Delta_bar'], bar_width2, color="0.8")
ax2.bar(fig10_h.index.values - 0.5*bar_width2, fig10_h['Delta'], bar_width2, color="0.4")
ax2.set_xlim(0, 6)
ax2.set_ylim(0, 20)
ax2.set_xticks([1, 3, 5])
ax2.set_xlabel("h")

bar_width3 = 6
ax3 = plt.subplot(gs[2])
ax3.bar(fig10_b.index.values - 0.5*bar_width3, fig10_b['Delta_bar'], bar_width3, color="0.8")
ax3.bar(fig10_b.index.values - 0.5*bar_width3, fig10_b['Delta'], bar_width3, color="0.4")
ax3.set_xlim(4, 34)
ax3.set_ylim(0, 20)
ax3.set_xticks([9, 19, 29])
ax3.set_xlabel("p")

plt.show()
```

```python
# save to pdf
pp = PdfPages('Figure10.pdf')
pp.savefig(fig10)
pp.close()
```
